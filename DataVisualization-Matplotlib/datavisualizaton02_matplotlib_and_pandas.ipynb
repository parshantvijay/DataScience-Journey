```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
countries = list(gas.columns)
countries = countries[1:]
countries
```

```python
plt.title("Gas Prices in Countries by Year")
plt.plot(gas['Year'],gas['USA'])
```

```python
#Plotting the Line Graph
for c in countries:
    plt.plot(gas['Year'], gas[c],'*-' ,label = c)

#Titles and Labels
plt.title("Gas Prices in Countries by Year",fontdict={'fontstyle':'normal','fontsize': 16,'fontweight': 'bold'})
plt.xlabel("Year")
plt.ylabel("Prices ($)")

plt.xticks(gas['Year'][::3]) #skipping to every 3 years

plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1))

plt.savefig('GasPrices_LineGraph.png',dpi = 300)

plt.figure(figsize=(15,8),dpi=500)
plt.show()
```

### FIFA DATA

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

```python
#Histogram 
bins = np.arange(40,110,10)
print(bins)

plt.hist(fifa['Overall'],bins=bins ,color='purple') # can also use hexadecimal for color

plt.xticks(bins)

plt.ylabel("Number of the Players")
plt.xlabel("Skill Level")
plt.title("Distribution of Player's Overall Skill Level")
plt.show()
```

### Pie Chart

```python
fifa.head()
```

```python
list(fifa.columns)
```

```python
fifa['Preferred Foot']
```

```python
# Percentage of people playing with left foot or right foot
leftfoot = fifa.loc[fifa['Preferred Foot'] =='Left'].count()[0]
print("Left Foot Count: ", leftfoot)

rightfoot  = fifa.loc[fifa['Preferred Foot'] =='Right'].count()[0]
print("Right Foot Count: ",rightfoot)

```

```python
labels = ['Left','Right']
colors = ['blue','green']
plt.pie([leftfoot,rightfoot],labels=labels,colors=colors, autopct='%.2f%%')
# first we gave the values to plot in the list
# second we gave the labels, third we gave the colors (both as a list)
# at last to show percentages we use autopct = '%.2f' to show percentages and %% to show the percentage sign
plt.title("Preffered Foot")
plt.show()
```

```python
##Breaking down the weights 
fifa['Weight']
```

```python
fifa['Weight'] = [x.strip('lbs')  if type(x)==str else x for x in fifa['Weight']]
fifa['Weight'][0]
fifa['Weight'] = pd.to_numeric(fifa['Weight'])
```

```python
light = fifa.loc[fifa['Weight'] < 125].count()[0]
print("Light Weight Count: ",light)

light_medium = fifa.loc[(fifa['Weight']>=125) & (fifa['Weight']<150)].count()[0]
print("light_medium Count: ",light_medium)

medium = fifa.loc[(fifa['Weight']>=150) & (fifa['Weight']<175)].count()[0]
print("Medium Count: ",medium)

medium_heavy = fifa.loc[(fifa['Weight']>=170) & (fifa['Weight']<200)].count()[0]
print("Medium_heavy Count: ",medium_heavy)

heavy = fifa.loc[fifa['Weight']>=200].count()[0]
print("Heavy Count: ",heavy)

weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['x < 125','125 <= x < 150', '150 <= x < 175','175 <= x < 200','x > 200']
explode = (.4,.2,0,0,.4)
plt.pie(weights, labels =labels,autopct='%.2f%%', pctdistance=0.8,explode=explode)

plt.title("Weight Distribution (lbs) ")
plt.figure(figsize=(200,200),dpi=300)
plt.show()

# NOTES: explode is used to give spaces between the parts in the graph. (in the same order as labels and weights)
```
