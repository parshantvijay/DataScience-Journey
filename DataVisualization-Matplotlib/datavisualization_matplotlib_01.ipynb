```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Basic Graphs

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)

plt.xlabel('x-axis')
plt.ylabel('y axis')
plt.title("First Graph")

plt.show()
```

```python
plt.plot(x,y)

plt.xlabel('x-axis', fontdict={'fontname': 'Times New Roman','fontsize': 16})
plt.ylabel('y axis',fontdict={'fontname': 'Times New Roman','fontsize': 16})
plt.title("First Graph", fontdict={'fontname': 'Times New Roman', 'fontsize' : 20})

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)

plt.xlabel('x-axis')
plt.ylabel('y axis')
plt.title("First Graph")

#changing the tick values to integers
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

```python
#Adding Legend 
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label = '2x') # Adding label to the line and showing legend

plt.xlabel('x-axis')
plt.ylabel('y axis')
plt.title("First Graph")

#changing the tick values to integers
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
#Changing colors
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label = '2x',color = 'green',linewidth = 1,marker = 'p', markersize = 12,markeredgecolor = 'blue') 
#Can also replace with hexa of colors
# linewidth changes the width of the line on the graph
# marker = '.','x','o' 'p'(pentagon)
# marker 

plt.xlabel('x-axis')
plt.ylabel('y axis')
plt.title("First Graph")

#changing the tick values to integers
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
#Changing colors
x = [0,1,2,3,4]
y = [0,2,4,6,8]


#Short hand notation
#fmt = '[color][maker][line]
plt.plot(x,y, 'g*--',label = '2x') 


plt.xlabel('x-axis')
plt.ylabel('y axis')
plt.title("First Graph")

#changing the tick values to integers
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
x2 = np.arange(0,4,0.5) # a range of values from 0 to 4 with 0.5 interval
x2
```

```python
plt.plot(x,y, 'g*--',label = '2x') 
plt.plot(x2,x2**2, 'b*-',label='Quadratic Function')

plt.legend()
plt.show()
```

```python
#Changing line style of the same line graph
x = np.arange(0,4.5,0.5)
plt.plot(x[:4],x[:4]**2, 'r*-',label='Quadratic Function') # till 3
plt.plot(x[3:],x[3:]**2, 'b*--',label='Quadratic Function') # from 4 till end
plt.legend()
plt.show()
```

```python
##Resizing the graph
plt.figure(figsize=(8,10),dpi=500) # figsize = (x,y) and dpi = pixel density(pixel per inch), use >=300 for a better picture quality
plt.plot(x,x**2,'r*--',label = 'y = x^2')
plt.legend()
plt.show()
```

```python
#Saving the Graph
##Resizing the graph
plt.figure(figsize=(8,10),dpi=500) # figsize = (x,y) and dpi = pixel density(pixel per inch), use >=300 for a better picture quality
plt.plot(x,x**2,'r*--',label = 'y = x^2')
plt.legend()
plt.savefig('mygraph.png',dpi = 300) # will save it in the current directory.
plt.show()

# Note: should use the savefig before show in order to save the graph
```

### BAR CHARTS

```python
labels  = ['A','B','C']
values = [1,4,2]

#Plotting Bar Graph
bargraph = plt.bar(labels,values)

patterns = ['/', 'O','*']

for bar in bargraph:
    bargraph[0].set_hatch(patterns.pop(0))
    
# bargraph[0].set_hatch('/')
# bargraph[1].set_hatch('O')
# bargraph[2].set_hatch('*')

#Labels
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.title('Bar Graph')

#Figure size
plt.figure(figsize=(6,4))

plt.show()
```
