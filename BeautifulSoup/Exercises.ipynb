```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://keithgalli.github.io/web-scraping/"
r = requests.get(url + "webpage.html")
webpage = bs(r.content)

print(webpage.prettify())
```

### Task_01. Grab all the social links from the webpage

```python
socials = webpage.select("ul.socials a")
socials
actual_links = [link['href'] for link in socials]
actual_links
```

```python
socials = webpage.find("ul",attrs={"class": "socials"})
links = socials.find_all("a")
actual_links = [link['href'] for link in links]
actual_links
```

```python
links = webpage.select("li.social a")
actual_links = [link['href'] for link in links]
actual_links
```

### Scraping the table on the webpage

```python
table = webpage.find("table",attrs={"class": "hockey-stats"})
print(table.prettify())
```

```python
table = webpage.select("table.hockey-stats")
table = table[0]
table
```

```python
import pandas as pd
table = webpage.select("table.hockey-stats")[0]
columns = table.find_all("th")
columns
```

```python
column_name = [c.string for c in columns]
column_name
```

```python
table_rows = table.find("tbody").find_all("tr")
l=[]
for tr in table_rows:
    td = tr.find_all('td')
    row = [str(tr.get_text()).strip() for tr in td]
    l.append(row)

df= pd.DataFrame(l,columns=column_name)
df.head()
```

### Grab all the funfacts that have the word 'is'

```python
funfacts = webpage.select("ul.fun-facts li")
funfacts
```

```python
import re
funfact_is = [fact.find(string=re.compile("is")) for fact in funfacts]
funfact_is = [fact.find_parent().get_text() for fact in funfact_is if fact]
funfact_is
```

### Solving the Mystery Challenge

```python
mystery = webpage.select("div.block a")
mystery
```

```python
relative_files = [f['href'] for f in mystery]
relative_files
```

```python
file1_url = relative_files[0]
full_url = url + file1_url
full_url
```

```python
for f in relative_files:
    full_url = url + f
    page = requests.get(full_url).content
    bs_page = bs(page)
    secret_word = bs_page.find("p", attrs= {"id":"secret-word"}).string
    print(secret_word)
```
