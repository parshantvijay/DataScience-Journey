### Reading Data in Pandas

```python
# Read Column names
pokemon.columns
```

```python
# Reading Specific Column
pokemon['Name']  # same can be achieved by "pokemon.Name"

# However it cannot be used for two words.
# Such as we cannot print Sp. Attk with dot method, we will have to use the square bracket method
#pokemon['Sp. Atk']

# Reading first entry of the Name Column
print(pokemon['Name'][0])

# Reading first 5 enteries of the Name Column
print(pokemon['Name'][0:5])
```

```python
# We can get multiple columns 
# By providing the list of Column Names
pokemon[['Name', 'Attack', 'Defense']]
```

```python
# Reading each Row
# we use iloc function to access the row at a specific index

pokemon.iloc[1] # Returns the row entry at the specific index

# Can print multiple rows too
pokemon.iloc[0:4] # Printing first 5 rows 
```

```python
# Reading a specific location (Row,Column)
pokemon.iloc[3,1]  
```

```python
# Iterating through each row in the dataset
for index, row in pokemon.iterrows():
    print(index,row['Name']) # Getting names in each row by iterating 
```

```python
#Finding specific data which is based on textual or numerical information
pokemon.loc[pokemon['Name']=='Charizard']
print("\n")
pokemon.loc[pokemon['Type 2']=='Poison']
```

### Sorting/Describing Data

```python
pokemon.describe()
```

```python
#Sorting values alphabetically 
pokemon.sort_values("Name")

# Sorting Values Alphabetically and printing in descending order
pokemon.sort_values("Name",ascending=False)

#Sorting Values on multiple columns and ascending on on column and descending the other
pokemon.sort_values(["Type 1","HP"],ascending=[True,False])
pokemon.sort_values(["Type 1","HP"],ascending=[1,0]) # Will give the same output as above
```

### Making Changes to the Data

```python
#Adding a column
pokemon['Total'] = pokemon['HP'] + pokemon['Attack'] + pokemon['Defense'] + pokemon['Sp. Atk'] + pokemon['Sp. Def'] + pokemon['Speed']
pokemon
```

```python
## Drop/Deleting a specific column
pokemon.drop(columns=['Total'],inplace=True)
pokemon
```

```python
##Adding a column in a different way
pokemon['Total'] = pokemon.iloc[:,4:10].sum(axis=1)
                    # : means all rows, 4:10 means column 4 to column 9 (last is exclusive)
                    #  (starts counting from 0)
                    # .sum() means sum all the values in those columns and axis = 1 means horizontally
                    # axis = 0 means vertically
pokemon
```

```python
## Re arranging columns

#moving Total Column 
columns = list(pokemon.columns.values)
print(columns)

pokemon = pokemon[columns[0:4] + [columns[-1]] + columns[4:(len(columns)-1)]]
pokemon
```

### Filtering Data

```python
pokemon.head()
```

```python
pokemon.loc[(pokemon['Type 1']=='Grass') & (pokemon['Type 2']=='Poison')]
```

```python
pokemon.loc[(pokemon['Type 1']=='Grass') | (pokemon['Type 2']=='Poison')]
```

```python
new_fd = pokemon.loc[(pokemon['Type 1']=='Grass') & (pokemon['Type 2']=='Poison') &(pokemon['HP'] > 70)]
new_fd.drop(columns='#',inplace=True)
new_fd.reset_index(drop=True,inplace=True)
new_fd
```

```python
new_fd.to_csv('new_filtered_data.csv',index=False)
```

```python
#List down all the pokemons having Mega in their names
pokemon.loc[pokemon['Name'].str.contains('Mega')]

#List down all the pokemons except Mega in their names
pokemon.loc[~pokemon['Name'].str.contains('Mega')]
```

```python
#Regex Filtering 
import re

pokemon.loc[pokemon['Type 1'].str.contains('(F|f)ire|(G|g)rass',regex=True)]
pokemon.loc[pokemon['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)]
```

```python
#Getting all pokemons name starting with pi
pokemon.loc[pokemon['Name'].str.contains('^pi[a-z]*',flags = re.I,regex=True)]
# ^ is used to specify ke bhai mujhe starting mein he pi chahiye beech mein nahi
```

### Conditional Changes

```python
pokemon.loc[pokemon['Type 1'] =='Fire','Type 1'] = 'Flamer'
```

```python
pokemon.loc[pokemon['Type 1']=='Flamer']
```

```python
pokemon.loc[pokemon['Type 1'] =='Flamer','Type 1'] = 'Fire'
pokemon
```

```python
#Make all the fire pokemon legendary
pokemon.loc[pokemon['Type 1']=='Fire','Legendary'] = 'True'

pokemon.loc[pokemon['Type 1']=='Fire']
#usecase: Using one condition to set the parameter of another column/feature 
```

```python
pokemon.loc[pokemon['Total']>500, ['Generation','Legendary']] = 'TEST VALUE' # setting TEST VALUE for both columns
pokemon
```

```python
pokemon.loc[pokemon['Total']>500, ['Generation','Legendary']] = ['TEST VALUE1','TEST VALUE2'] # setting TEST VALUE for both columns
pokemon
```

### Aggregate Statistics

```python
pokemon.groupby(["Type 1"]).mean(numeric_only=True)
```

```python
pokemon.groupby('Type 1').count().sort_values('Defense',ascending=False)
```

```python
pokemon.groupby(['Speed']).count()
```

```python
pokemon['Count'] = 1
pokemon
```

```python
pokemon.groupby(['Type 1']).count()['Count']
```

```python
# Can use multiple parameters in groupby()
pokemon.groupby(['Type 1','Type 2']).count()['Count'].reset_index()
```

### Working with large amount of data

```python
new_df = pd.DataFrame(columns=pokemon.columns)

for df in pd.read_csv('Modified_Pokemon.csv',chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])

new_df
```
